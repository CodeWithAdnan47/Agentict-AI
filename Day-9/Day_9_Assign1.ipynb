```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=userdata.get('Assignments2'))
```

```python
loader = PyMuPDFLoader("story.pdf")
documents = loader.load()
full_text = "\n".join(doc.page_content for doc in documents)
splitter = RecursiveCharacterTextSplitter(
    chunk_size=30,
    chunk_overlap=0
)

chunks = splitter.split_text(full_text)
translate_prompt = PromptTemplate( input_variables=["text"],
    template="""
Translate the following English text into Hindi.
Keep the translation clear and natural.
Do not add explanations.

English Text:
{text}
"""

)

translation_chain = translate_prompt | llm | StrOutputParser()
for i, chunk in enumerate(chunks, start=1):
    hindi_translation = translation_chain.invoke({"text": chunk})

    print(f"Chunk {i} (EN): {chunk}")
    print(f"Translation (HI): {hindi_translation}")
```
