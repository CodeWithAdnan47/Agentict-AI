✅ Problem 1: Collect Student Details (Variables + Input + List)
Write a Python program that:

Takes the student name as input
Takes marks for 3 subjects as input
Stores the marks inside a list called marks
Prints the name and marks list

```python
#Problem-1
name = input("Enter Name of The Student: ")
marks=[]

for i in range(3):
  sub1 = int(input(f"Enter Marks of Subject{i+1}: "))
  marks.append(sub1)
print(f"Name of the Student: {name}")
print(f"Marks of the Student: {marks}")

```

✅ Problem 2: Calculate Total and Average (Loop + List)
Using the marks list from Problem 1:

Use a loop to calculate the total marks
Calculate the average marks
Print both total and average

```python
#Problem-2
for i in marks:
  total_marks = sum(marks)
  avg_marks = total_marks/len(marks)
print(f"Total Marks of {name}: {total_marks}")
print(f"Average Marks of {name}: {avg_marks}")

```

✅ Problem 3: Decide Pass or Fail (Conditionals + Loop)
Using the marks list:

Check if the student has scored 35 or above in all subjects
If yes → print "Pass"
If no → print "Fail"

```python
#problem-3
counter = 0
result = "Pass"

for i in marks:
  counter+=1
  if i>=35:
    print(f"Student{name} Passed in Subject{counter}")
  else:
    result = "Fail"
    print(f"Student {name} Failed in Subject {counter}")
print(f"Main Result: {result}")
```

Create a dictionary called student that stores:
