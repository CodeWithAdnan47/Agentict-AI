Question-1: Smart Greeting Function for a GenAI App
Write a function named generate_greeting(name) for a GenAI assistant app.

Requirements:

Accept one positional argument: name (string).
If the user gives extra spaces (example: " chandu "), your function must clean the name using strip().
If the cleaned name is an empty string (example: " "), return: "Hello! Welcome to GenAI."
Otherwise, return: "Hello, <Name>! Welcome to GenAI." where <Name> must be title-cased using .title().
Test cases you must run:

generate_greeting("Chandu")
generate_greeting(" akhila ")
generate_greeting(" ")

```python
#Problem-1

def greetings(name):
  name = name.strip()
  if name=="":
    return "Hello!! Welcome To GenAI!!"
  else:
    return f"Hello,{name.title()}! Welcome To GenAI!!"
print(f"Test Case 1: {greetings("Chandu")}")
print(f"Test Case 2: {greetings(" akhila ")}")
print(f"Test Case 3: {greetings(" ")}")
```

Question-4: return vs print — Build a Reusable Discount System
Create two functions for a GenAI subscription discount.

Function-1: discount_print(price, discount_percent)

Calculates discounted price and prints it
Does not return anything
Function-2: discount_return(price, discount_percent)

Calculates discounted price and returns it
Does not print anything
Formula: final = price - (price * discount_percent / 100)

Proof code (must do):

a = discount_print(1000, 10)

b = discount_return(1000, 10)

Print a and b

Add comments explaining:

why a is None
why b is 900

```python
#Problem-4
def discount_print(price, discount_percent):
  discounted_price = price - (price * discount_percent / 100)
  print(f"Discounted Price: {discounted_price}")
def discount_return(price, discount_percent):
    discounted_price_2 = price - (price * discount_percent / 100)
    return f"Discounted_Price_2: {discounted_price_2}"
a = discount_print(1000, 10)
#discount_print() doesn't return any it just simply print it so bcz it doesn't returning any value therefore 'a' prints None
b = discount_return(1000, 10)
#discount_return() uses return keyword to show/print the price so it directly return the result not prints it
print(a)
print(b)

```

Question-5: Lambda Tasks for Prompt Processing
Given:

prompts = ["Write a poem", "Summarize", "Generate code", "Hi", "Explain AI", "Ok"]
Tasks:

Use map() + lambda to create a list of lengths of each prompt.
Use filter() + lambda to create a list of prompts with length > 5.
Print both lists.

```python
#Problem-5
prompts = ["Write a poem", "Summarize", "Generate code", "Hi", "Explain AI", "Ok"]
prompt_length = list(map(lambda m: len(m),prompts))
filtered_prompts = list(filter(lambda m: len(m)>5,prompts))
print(f"The List of Lenghth of Each Prompt: {prompt_length}")
print(f"The List of Filtered Prompts: {filtered_prompts}")
```

Write exactly:

Step-2: Read Config

Open in read mode
Read full content using .read()
Print it
Close the file
Step-3: Verify Close

Print f.closed before closing
Print f.closed after closing


```python
#PROBLEM-6
with open('config.txt','w') as f:
  #Writing Content
  f.write('\n Model=ChatGPT')
  f.write('\n Temperature=0.7')
  f.write('\n Max_Tokens=200')
  print("Write Operation is Succesfull!!")
  print(f"Closing The File Before Performing Read Operation.{f.closed}")
  f.close()
with open('config.txt','r') as f:
  content = f.read()
  print("\n Content Present in 'config.txt': ")
  print(content)
  print(f"Closing The File After Performing Read Operation.{f.closed}")
  f.close()
  print("Read Operation is Succesfull!!")
```

Question-7: CSV File Handling — Filter Logs by Token Usage
You have model_logs.csv with columns: Timestamp,Prompt,Tokens,ResponseTime

Task:

Use csv.reader
Skip header using next(reader)
Print only rows where Tokens > 100
Count such rows and print:
High Token Requests: <count>
Important: Convert token strings to int.

```python
#Problem-7
import csv
import os
dummy_csv =  [
        ['Timestamp', 'Prompt', 'Tokens', 'ResponseTime'],
        ['2023-10-27 10:00:00', 'Generate a short story', '150', '0.5'],
        ['2023-10-27 10:01:15', 'Summarize article X', '80', '0.2'],
        ['2023-10-27 10:02:30', 'Write a poem about nature', '220', '0.8'],
        ['2023-10-27 10:03:45', 'Translate phrase Y', '50', '0.1'],
        ['2023-10-27 10:05:00', 'Explain quantum physics', '300', '1.2'],
        ['2023-10-27 10:06:15', 'Tell a joke', '20', '0.05'],
        ['2023-10-27 10:07:30', 'Create code for sorting list', '180', '0.7']
    ]
with open('model_logs.csv','w',newline='') as f:
  writer = csv.writer(f)
  writer.writerows(dummy_csv)
  print("CSV File Created Successfully!!")
  high_token_requests = 0
with open('model_logs.csv','r') as f:
  reader = csv.reader(f)
  next(reader)
  for row in reader:
    if int(row[2]) > 100: #tokens are at the 3rd position in file i.e index==2
      print(row)
      high_token_requests += 1 #incrementing the count of token request count
print(f"High Tokens Request: {high_token_requests}")
```

Question-9: Exception Handling — Safe Calculator for GenAI Tokens
Write a program that takes:

total_tokens (input)
num_requests (input)
Compute: avg_tokens = total_tokens / num_requests

Requirements:

Handle ValueError for non-numeric input
Handle ZeroDivisionError for 0 requests
If valid, print:
Average Tokens Per Request: <avg_tokens>
Mandatory Bonus:

Add except Exception as e to catch unknown errors and print:
Unexpected error: <error_message>

```python
#Problem-9
try:
  total_tokens = int(input("Enter Total Tokens: "))
  num_req = int(input("Enter No. of Requests: "))
  avg_tokens = total_tokens / num_req #Error if User Entered 0
  print(f"Average Tokens Per Request: {avg_tokens}")
except ValueError:
  print("Invalid Input!! Please Enter Integer(Numeric) Value")
except ZeroDivisionError:
    print("Error: Division by 0 is Not Allowed.")


```
